```python
import pandas as pd
import ast
import numpy as np
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
part1 = pd.read_csv("csv_results/result_32.csv")
part2= pd.read_csv("csv_results/result_33.csv")
part1
```

```python
theta_0_lst = [0.2, 0.5, 0.8]
theta_1_lst_1 = [0, 0.2]
theta_1_lst_2=[0.5, 0.8]
model_lst = ["ARMA","Informer","RNN"]
DGP_lst = ["arma", "FAR", "bilinear"]
```

```python
def getMSE(DGP_name,model,theta_0,theta_1,output):
    EX = np.array(output[f"{DGP_name}_{theta_0}_{theta_1}_True"].apply(ast.literal_eval).to_list())

    return mean_squared_error(
            EX,
            np.array(output[f"{DGP_name}_{theta_0}_{theta_1}_{model}"].apply(ast.literal_eval).to_list()),
            multioutput="raw_values",
        )[0]
```

```python
result = {}
for DGP in DGP_lst:
    for model in model_lst:
        column = (DGP,model)
        temp_result= {}
        for theta_0 in theta_0_lst:
            for theta_1 in theta_1_lst_1:
                combination = (theta_0,theta_1)
                temp_result[combination]=getMSE(DGP_name=DGP,model=model,theta_0=theta_0,theta_1=theta_1,output=part1)
        result[column] = temp_result
result            
```

```python
for DGP in DGP_lst:
    for model in model_lst:
        column = (DGP,model)
        for theta_0 in theta_0_lst:
            for theta_1 in theta_1_lst_2:
                combination = (theta_0,theta_1)
                result[column][combination]=getMSE(DGP_name=DGP,model=model,theta_0=theta_0,theta_1=theta_1,output=part2)
```

```python
result
```

```python
df = pd.DataFrame(result)
df.columns = pd.MultiIndex.from_tuples(df.columns, names=['DGP','Model'])
```

```python
df.sort_index(inplace=True)
print(df)
```

```python
df.to_csv("table1.csv")
```
